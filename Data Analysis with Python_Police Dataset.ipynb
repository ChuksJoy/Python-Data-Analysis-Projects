### Police Dataset

This is a Simple Data analysis of a data from a Police Check Post using the Pandas Dataframe


#### Importing Our Dataframe and Dataset

```python
import pandas as pd
```

```python
data = pd.read_csv('Police Data.csv')
```

```python
data
```

## Data Analysis


#### 1. Data Cleaning

Remove the Column that only contains missing values/ null values

```python
data.isnull()
```

```python
# checking to see the sum of all null values

data.isnull().sum()
```

```python
# removing the column with null values
data.drop(columns = 'country_name', inplace = True)
```

```python
data
```

#### 2. Filtering and Value Count

For Speeding, were Men or Women Stopped more Often?

We will be considering the Violation column and the driver_gender column

```python
data.head(3)
```

```python
data[data.violation == 'Speeding']
```

```python
data[data.violation == 'Speeding'].driver_gender.value_counts()
```

#### 3. Groupby

Does gender affect who gets searched during a stop?
We will be using the search_conducted and driver_gender Column

```python
data.head(2)
```

```python
data.groupby('driver_gender').search_conducted.sum()
```

```python
# how many times search was conducted on both gender

data.search_conducted.value_counts()
```

#### 4. Mapping and Data-type Casting

What is the mean stop_duration?

```python
data.head(2)
```

```python
# value count of the stop_duration column

data.stop_duration.value_counts()
```

```python
# converting the datatype of the column to an integer type

data['stop_duration'] = data['stop_duration'].map( {'0-15 Min' : 7.5, '16-30 Min' : 24, '30+ Min' : 45 })
```

```python
data
```

```python
data['stop_duration'].mean()
```
