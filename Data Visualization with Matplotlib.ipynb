#### Ploting Our Line Graph

```python
# Basic Line graph for our data

plt.plot(gas.Year, gas.Australia);
plt.plot(gas.Year, gas.USA);
plt.plot(gas.Year, gas['South Korea']);
```

```python
# Modifying the graph to look better

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
plt.plot(gas.Year, gas.Canada, 'r.-')
plt.plot(gas.Year, gas['South Korea'], 'g.-')
plt.plot(gas.Year, gas.Australia, 'y.-')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

```python
# Alternative way of plotting the above graph
countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:
    if country in countries_to_look_at:
         plt.plot(gas.Year, gas[country], marker='.', label = country)
            
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})
plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

```python
#visualizing all countries
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

for country in gas:
    if country != 'Year':
         plt.plot(gas.Year, gas[country], marker='.', label=country)# add a label for each plot line

            

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()# create the legend using the labels


plt.show()
```

```python
# moving the label outside the chart using bbox_to_anchor
plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

for country in gas:
    if country != 'Year':
         plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # specify the position of the legend

plt.show()
```

##### Pie Chart

```python
# looking at the dataset and filtering the dataset to see players that prefer their left foot

left = fifa.loc[fifa['Preferred Foot']== 'Left'].count()[0]#count gives us an array, but assing a number 0 gives us a number

left
```

```python
# right foot
right = fifa.loc[fifa['Preferred Foot']== 'Right'].count()[0]

right
```

```python
# trying to plot the prefernce of players preferred foot

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
color = ['orange', 'blue']

plt.pie([left, right], labels = labels, colors = color);
```

```python
#modifying the pie chart

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['yellow', 'purple']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
# plotting another pie chart with the break down of the weights of the players
#first lets look at what the weight is like

fifa.Weight
```

```python
# stripping off the lbs in the output to just have values

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')


light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')# explode breaks the graph apart
plt.show()
```

#### Box and Whiskers Chart

Will create a box plot comparing the overall ratings of three professional soccer teams: FC Barcelona, Real Madrid, and New England Revolution from our fifa dataset.

The box plot is created using the matplotlib library and includes three boxes, one for each team, with the overall ratings of players in the team represented as individual points within the box. The boxes will be  colored with a light gray color and have blue-colored edges. The median line in each box is given a thickness of 2.

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
